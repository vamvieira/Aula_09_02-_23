```python
#Importando a biblioteca OpenCV
import cv2
#Impotando a biblioteca Numpy8 bits(realizar operações com matrizes)
import numpy as np
```

```python
#Linha magica para imprimir graficos no notebook
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
print("OpeCV Versão:%s" %cv2.__version__)
```

```python
image = cv2.imread ("scale.jpeg")
plt.imshow(image)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)

plt.axis('off')
plt.show()
```

```python
image2 = cv2.imread ("NATUREZA_1.jpg")
plt.imshow(image2)
```

```python
image2_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
plt.imshow(image2_rgb)

plt.axis('off')
plt.show()
```
